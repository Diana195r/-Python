Как заменять элементы строки? Для замены элементов в строке используется метод replace.

str.replace(old, new[, count])

Здесь

    str — строка, к которой применяется метод;
    old — подстрока, которую необходимо найти и заменить;
    new — новая подстрока, которой будет заменена старая;
    count — необязательный аргумент, целое число - количество совпадений старой подстроки, которую необходимо заменить; если параметр опущен, то заменеютя все вхождения.

Результат - новая строка.


````python
#пример использования replace.
s=s.replace('умножения', 'сложения')
print(s)
````

**Срезы в Python.** 
Язык Python позволяет обратиться к фрагменту строки с помощью такого объекта, как срез. Оператор извлечения среза из строки  имеет вид
```
S[n:k:d],
```
здесь 
* n – начальный индекс строки, по умолчанию – 0,
* k – конечный индекс строки, по умолчанию – len(S), конечный индекс в срез не входит
* d – шаг изменения индекса, по умолчанию – 1. 

Отрицательный индекс соответствует нумерации в обратном порядке, **-1** - *последний* символ в строке, **-2** - *предпоследний*, ..., **-len(s)** - это *нулевой* элемент строки (**-n** равен **len(s)-n**). 

````python
s='примерчик'
print(s[::])
print(s[:len(s):])
print(s[-1:-len(s)-1:-1])
````

Оператор in определяет принадлежит ли подстрока строке.

````python
s="Василий Всеволодович"
s1="Василий"
print(s1 in s)
print(s in  s1)
````

Функции min и max применительно к строкам определяют символ с минимальным (максимальным) кодом символа.

````python
print(min("примерчик"),max("примерчик"));
````

Лишние пробелы в строку представляют определённую проблему при программировании. Python позволяет удалить лишние пробелы из строк. Метод rstrip() удаляет лишние пробелы справа, а метод lstrip() – слева.

````python
x="    готовимся к экзамену по информатике     "
print(x.title().lstrip().rstrip()+"!")
print(x.title()+"!")
````

Метод strip() удаляет пробелы с обоих концов. Но как видно, из примеров, методы не модифицируют исходную строку. Чтобы строка изменилась, необходимо результат методов переписать в переменную.

````python
x="     готовимся к экзамену по информатике     "
x=x.title().strip()
print(x+"!")
````

#  Списки в Python

*Список* – это набор элементов, следующих в определенном порядке. В список можно включать любые данные, которые даже не обязаны быть как-то связаны друг с другом. Список *обычно* содержит более одного элемента.

````python
spis = ['Русский язык', 'История', 'Математика', 'Информатика']
print(spis)
````

Чтобы обратиться к элементам списка, необходимо указать имя списка и в квадратных скобках его номер. При выводе списка, он выводится в квадратных скобках. 

````python
print(spis[0],spis[2])
print(spis[1].upper())
````

Элементами списка могут быть значения любого типа, в том числе числа и другие списки.

````python
spisok=[1,5.6,['Информатика','Программирование'],[3,4,5,8+7j]];
print(spisok)
````

Над списками, как и над строками, определены операции + и *.

````python
spisok=[1,5.6,['Информатика','Программирование'],[3,4,5,8+7j]];
spisok1=[13,14,'Математика'];
spisok2=spisok+spisok1;
spisok4=spisok1*15;
print(spisok,'\n',spisok1,'\n',spisok2,'\n',spisok4);
````

## Изменение, добавление и удаление элементов списка

Списки в Python являются динамическими, поэтому актуальным является вопрос модификации списка.
В отличии от строки, элемент списка можно изменить обычным оператором присваивания.

````python
spis = ['Русский язык', 'История', 'Математика', 'Информатика']
spis[1]='Литература';
print(spis);
````

Однако, добавить элемент в список обычным присваиванием не получится.
В списках также можно использовать срезы.
Если мы хотим получить строку с большей гибкостью обработки, то можно создать список из символов.


**Добавление элементов в список**

Python предоставляет несколько способов добавления новых данных в существующие списки. Для присоединения элемента в конец списка используется метод **append(list)**, здесь *list* – добавляемый в список элемент.


````python
spis = ['Русский язык', 'История', 'Математика', 'Информатика']
print(spis)
spis.append('История')
print(spis)
````

Можно начать с пустого списка, добавляя в него элементы, используя метод **append()**.

````python
spis=[]; 
spis.append('Русский язык'); 
spis.append('История');
spis.append('Математика'); 
spis.append('Информатика');
spis.append('Музыка') 
print(spis)
````

Метод **insert()** позволяет добавить новый элемент в произвольную позицию списка. Для этого следует указать индекс и значение нового элемента. 

````python
spis = ['Русский язык', 'Математика', 'Информатика']
print(spis)
spis.insert(2,'История')
print(spis)
````

Добавление *списка* в конец  другого *списка* .

````python
spis = ['Русский язык', 'Математика', 'Информатика']
spis2=['Литература','Музыка'];
spis.extend(spis2);
print(spis);
````

**Удаление элементов из списка**
Нередко возникает необходимость в удалении одного или нескольких элементов из списка. Элементы удаляются из списка по позиции или по значению.  

*Удаление элемента с использованием команды del*

С помощью команды *del* можно удалить элемент, если известна его позиция в списке. Синтаксис её следующий 
```
del list[i]
```

````python
spis = ['Русский язык', 'Математика', 'Информатика', 'Литература', 'Музыка']
del spis[1];
print(spis);
````

Ещё одним способом удаления значения из списка является метод **pop(i)**.
```
L.pop(i)
```
Этот метод извлекает из списка L элемент с номером i, после чего удаляет его, если номер i опущен (**L.pop()**), то извлекается и удаляется последний элемент списка.

````python
spis = ['Русский язык', 'Математика', 'Информатика', 'Литература', 'Музыка'] 
x=spis.pop(2);
print(x, '\n',spis)
x=spis.pop()
print(x, '\n',spis)
````

Удаление из списка первого элемента с заданным значением. Для выполнения заданной операции служит метод **remove**.
```
L.remove(x)
```
Здесь L – имя списка, x – удаляемое из списка значение.

````python
spisok=[1,5.6,3,4,5,8+7j,5.6,17,5.6];
spisok.remove(5.6)
print(spisok);
spisok.remove(5.6)
print(spisok);
spisok.remove(5.6)
print(spisok);
````

Метод **L.cout(x)** вычисляет количество элементов списка, равное х.

````python
spisok=[1,5.6,3,4,5,8+7j,5.6,17,5.6];
print(spisok.count(5.6))
````

Метод **L.index(x)** вычисляет индекс первого вхождения значения х в список L.

````python
spisok=[1,5.6,3,4,5,8+7j,5.6,17,5.6];
print(spisok.index(5.6))
````

Метод **L.sort()** располагает элементы списка L по возрастанию.

Метод **L.reverse()** располагает элементы списка L в обратном порядке.

````python
spisok=[-1,5,0.6,3,4,5,8,7,5.6,-17,-5.6];
spisok.sort()
print(spisok)
spisok.reverse()
print(spisok)
````

# Кортежи

Список так же может быть неизменяемым, как и строка, в этом случае он называется кортеж (**tuple**). Кортеж использует меньше памяти, чем список. При задании кортежа вместо квадратных скобок используются круглые (хотя можно и совсем без скобок). Кортеж не допускает изменений, в него нельзя добавить новый элемент, удалить или заменить существующие элементы, но он может содержать изменяемые объекты, например, списки:

````python
L1=[];
A=(1,3,-7,L1);
print(A);
````

````python
L1=[];
A=(1,3,-7,L1);
print(A);
A[1]=4;
````

Видно, что прямая замена элемента кортежа недопустима – вызывает **TypeError**, так как тип **tuple** не поддерживает изменение элементов, но если использовать втроенный метод append у списка, являющегося элементом кортежа, этот список можно изменить.

````python
L1=[];
A=(1,3,-7,L1);
print(A);
A[3].append(143)
print(A,L1)
````

Функция **tuple(s)** преобразовывает строку или список s в кортеж,  функция **list(s)** переводит кортеж s в список.

Основное различие между кортежами и списками состоит в том, что кортежи не могут быть изменены. На практике это означает, что у них нет методов, которые бы позволили их изменить: **append()**, **extend()**, **insert()**, **remove()**, **pop()**.
Но можно взять срез от кортежа, так как при этом создастся новый кортеж. Кортежи в некоторых случаях быстрее, чем списки. Кортежи делают код безопаснее в том случае, если у вас есть «защищенные от записи» данные, которые не должны изменяться. Некоторые кортежи могут использоваться в качестве элементов множества и ключей словаря (конкретно, кортежи, содержащие неизменяемые значения, например, строки, числа и другие кортежи).
В Python можно использовать кортежи, чтобы присваивать значение нескольким переменным сразу:

````python
v = ('f',5, True )
(x,y,z) = v
print(x)
print(y)
print(z)
````
